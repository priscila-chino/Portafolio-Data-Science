## 3. Limpieza de datos

A continuaci贸n se realizar谩 la limpieza del dataset basada en los problemas identificados durante la exploraci贸n inicial.

### 3.1. Manejo de Valores Nulos

Durante la exploraci贸n se identificaron valores nulos en las columnas "proveedor" (121 nulos) y "estado" (72 nulos).

#### Decisi贸n de tratamiento:
- proveedor: Se rellenar谩n los valores nulos con "Desconocido"
- estado: Se rellenar谩n los valores nulos con "Pendiente" (asumiendo que si no se especifica el estado, el pago est谩 pendiente)

```python
# Crear copia del dataframe para preservar los datos originales
df_clean = df.copy()

# Rellenar los valores nulos
df_clean["proveedor"] = df_clean["proveedor"].fillna("Desconocido")
df_clean["estado"] = df_clean["estado"].fillna("Pendiente")

# Verificar que ya no hay valores nulos
print("Verificaci贸n de valores nulos despu茅s del tratamiento:")
print(df_clean.isnull().sum())
```

### 3.2. Detecci贸n especifica y correcci贸n de Fechas Invalidas

Se observaron fechas con valores imposibles como "2024/05/32" (32 de mayo) y "2024/02/30" (30 de febrero).
Por ello necesitamos identificar todas las fechas invalidas, entonces utilizaremos el parametro `errors="coerce"` de `pd.to_datetime()`, que convierte las fechas invalidas en `NaT` (representa que un dato no es ni fecha ni hora). Luego se corregir谩n manualmente las fechas espec铆ficas que fallaron.


#### Decisi贸n de tratamiento:
Estas fechas se corregir谩n a los 煤ltimos d铆as v谩lidos de cada mes.

```python
print("----- DETECCION DE FECHAS INVALIDAS ------")

# Intentar convertir a datetime y capturar errores
fechas_prueba = pd.to_datetime(df_clean['fecha'], errors="coerce")
fechas_invalidas = df_clean[fechas_prueba.isna()] # Esta linea solo guarda los True

print(f"\nSe encontraron {len(fechas_invalidas)} fechas invalidas:")
print(fechas_invalidas["fecha"].unique())
```

```python
# Funcion para corregir fechas invalidas basada en la deteccion anterior
def corregir_fechas(fecha):
    fecha_str = str(fecha)
    if "2024/05/32" in fecha_str: return "2024-05-31"
    elif "2024/02/30" in fecha_str: return "2024-02-28"
    elif "2024/13/05" in fecha_str: return "2024-05-13"
    elif "2024/06/31" in fecha_str: return "2024-06-30"
    elif "2024/07/35" in fecha_str: return "2024-07-31"

    else: return fecha_str
# Aplicar correcci贸n
df_clean["fecha"] = df_clean["fecha"].apply(corregir_fechas)

# Convertir definitivamente a datetime
df_clean["fecha"] = pd.to_datetime(df_clean["fecha"], errors="coerce")
```

```python
# Despu茅s de cada paso, verificar con:
print(df_clean.info())
print(df_clean.head())
```

### 3.7. Validaci贸n de Reglas de Negocio - Relaci贸n Categor铆a/Tipo_Pago

Seg煤n las reglas de negocio definidas para la administraci贸n de consorcios:

**Regla 1:** Si `categoria = "Expensas Extraordinarias"`  `tipo_pago DEBE ser "Extraordinario"`

**Regla 2:** Para todas las otras categor铆as (`Limpieza`, `Mantenimiento`, `Seguridad`, `Impuestos`)  `tipo_pago DEBE ser "Ordinario"`

Estas reglas aseguran la consistencia l贸gica entre la naturaleza del gasto y su forma de financiamiento.

```python
# Detectar inconsistencias en las reglas de negocio
print("------ VALIDACION DE REGLAS DE NEGOCIO -----")

# Regla 1: Expensas Extraordinarias deben tener tipo_pago Extraordinario
error_regla1 = df_clean[(df_clean["categoria"] == "Expensas Extraordinarias") &
                        (df_clean["tipo_pago"] != "Extraordinario")]

# Regla 2: Otras categor铆as deben tener tipo_pago Ordinario
categorias_ordinarias = ["Limpieza", "Mantenimiento", "Seguridad", "Impuestos"]
error_regla2 = df_clean[(df_clean["categoria"].isin(categorias_ordinarias)) &
                        (df_clean["tipo_pago"] != "Ordinario")]

# Combinar todos los errores
errores_reglas = pd.concat([error_regla1, error_regla2])

print(f"Se encontraron {len(errores_reglas)} inconsistencias en las reglas de negocio:")

if len(errores_reglas) > 0:
    print("\nRegistros con inconsistencias:")
    display(errores_reglas[['id_gasto', 'categoria', 'tipo_pago', 'descripcion', 'monto']])

    # Resumen por tipo de error
    print("\nResumen de inconsistencias:")
    print("1. Expensas Extraordinarias con pago NO Extraordinario:", len(error_regla1))
    print("2. Categor铆as ordinarias con pago NO Ordinario:", len(error_regla2))
```
